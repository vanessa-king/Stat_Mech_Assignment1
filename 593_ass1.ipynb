```python
def b_tilda(n):
    a = 1
    b = 3
    c = 0
    return a*n**2 + b*n + c

def eq_of_state(num_terms_in_sum, eta):
    state = 1
    for i in range(1, num_terms_in_sum):
        state += b_tilda(i)*eta**i
    return state

def perfect_eq_of_state(eta):
    return 1 + 4*eta + 10*eta**2 + 18.365*eta**3 + 28.24*eta**4 + 39.5*eta**5 + 55.5*eta**6

def barboy_gelbart_eq(eta):
    return 1/(1-eta) + 3*eta/(1-eta)**2 + 3*eta**2/(1-eta)**3 - 0.635*eta**3/(1-eta)**4
```

```python
eta, step = np.linspace(0.001, 1, num=50, endpoint=True, retstep=True)

state = [eq_of_state(7,i) for i in eta]
perfect_state = [perfect_eq_of_state(i) for i in eta]
barboy_gelbart = [barboy_gelbart_eq(i) for i in eta]
```

```python
#Plotting: 
fig, ax = plt.subplots()

ax.plot(eta, state, label="Carnahan & Starling (7 terms)")
#ax.plot(eta, perfect_state, label="Numerical Integration results")
ax.plot(eta, barboy_gelbart, label="Barboy & Gelbart (4 terms)")

ax.set_xlabel(r'$\eta$')
ax.set_ylabel('Equation of state')
ax.set_ylim(-50, 150)
plt.legend()
plt.title("Equation of state value versus packing fraction")

plt.show()
```
